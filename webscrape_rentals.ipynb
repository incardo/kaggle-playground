```python
import requests
from bs4 import BeautifulSoup

def scrape_ft_articles():
    url = 'https://www.ft.com/'
    headers = {'User-Agent': 'Mozilla/5.0'}  # Some websites require a User-Agent header

    # Fetch the web page
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"Failed to fetch {url}")
        return []

    # Parse HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all article elements
    article_elements = soup.find_all('div', class_='o-teaser__content')  # Adjust class based on the website structure

    # Extract article titles and URLs
    articles = []
    for article in article_elements:
        title = article.find('span', class_='o-teaser__heading').text.strip()
        article_url = article.find('a')['href']
        articles.append({'title': title, 'url': article_url})

    return articles

# Example usage
ft_articles = scrape_ft_articles()
for article in ft_articles:
    print(article['title'])
    print(article['url'])
    print('-' * 50)
```

```python
import requests
from bs4 import BeautifulSoup
news_titles=[]

for page in range(1,6):
    url="https://www.ft.com/world?page={}".format(page)
    result=requests.get(url)
    reshult=result.content
    soup=BeautifulSoup(reshult, "lxml")
    for title in soup.findAll("div",{"class":"o-teaser__heading"}):
        titles=title.find(text=True)
        news_titles.append(titles)

#   <p class="o-teaser__standfirst"><a href="/content/7af8a288-2762-4846-a021-8ba161e003cd" data-trackable="standfirst-link" tabindex="-1" class="js-teaser-standfirst-link">We must act now to equip the next generation of builders, engineers and architects with skills to restore a ravaged country</a></p>

print(news_titles)
```
